# CycleGAN Example
Based on the work by xhujoy - https://github.com/xhujoy/CycleGAN-tensorflow

```python
import argparse
import os
import tensorflow as tf
```

```python
tf.set_random_seed(98052)
```

```python
# Import the CycleGAN implemenation
from model import cyclegan
```

```python
# Define the argument class
# Note: epoch is set to 200 in the original implementation
# We set it to 1 to enable readers to be able to go through the flow.

class args:
  dataset_dir='horse2zebra'
  epoch=1
  epoch_step=100
  batch_size=1
  train_size=1e8
  load_size=286
  fine_size=256
  ngf=64
  ndf=64
  input_nc=3
  output_nc=3
  lr=0.0002
  beta1=0.5
  which_direction='AtoB'
  save_freq=1000
  print_freq=100
  continue_train=False,
  checkpoint_dir='./checkpoint'
  sample_dir='./sample'
  test_dir='./test'
  L1_lambda=10.0
  use_resnet=True
  use_lsgan=True
  max_size=50
  _phase='train'

  @property
  def phase(self):
    return type(self)._phase

  @phase.setter
  def phase(self,val):
    type(self)._phase = val
    
  
```

```python
# Create the relevant directories
if not os.path.exists(args.checkpoint_dir):
  os.makedirs(args.checkpoint_dir)

if not os.path.exists(args.sample_dir):
  os.makedirs(args.sample_dir)

if not os.path.exists(args.test_dir):
   os.makedirs(args.test_dir)
```

```python
import pprint

pp = pprint.PrettyPrinter(indent=2)
pp.pprint(vars(args))
```

## Train the CycleGAN

```python
tfconfig = tf.ConfigProto(allow_soft_placement=True)
tfconfig.gpu_options.allow_growth = True

with tf.Session(config=tfconfig) as sess:
  model = cyclegan(sess, args)
  model.train(args) 
    
```

```python
# Un-comment this line to see the logs in TensorBoard
# !tensorboard --logdir=./logs
```

## Test the CycleGAN from A to B

```python
tfconfig = tf.ConfigProto(allow_soft_placement=True)
tfconfig.gpu_options.allow_growth = True

tf.reset_default_graph() 
args.phase='test'

with tf.Session(config=tfconfig) as sess:
  model = cyclegan(sess, args)
  model.test(args) 
    
```
